```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

## Creating DataFrame

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

## Add Target Column

```python
df['target'] = iris.target
df.head()
```

## Filter Data for a Specific Target

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

Add Flower Name Column

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
df.head()
```

## Prepare Data for Modeling

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'], axis='columns')
y = df.target
```

## Split the Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

Parameters:
test_size=0.2: 20% of the data is reserved for testing.
random_state=1: Ensures reproducibility.




```python
len(X_train)
```

len(X_train): Returns the number of training samples (120 in this case)

```python
len(X_test)
```

len(X_test): Returns the number of testing samples (30 in this case)

## Initialize the kNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
```

## Train the Model

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

## Make a Prediction

```python
input_data = pd.DataFrame([[4.8, 3.0, 1.5, 0.3]], columns=X.columns)

knn.predict(input_data)
```

## Generate the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

## Visualize the Confusion Matrix

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

## Display the Classification Report

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
